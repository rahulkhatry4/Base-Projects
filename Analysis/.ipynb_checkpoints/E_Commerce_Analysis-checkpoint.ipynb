# Data Cleanup

```python
# Droping NAs and faulty orders

customers_data = customers_data.dropna()
geolocation_data = geolocation_data.dropna()
order_items_data = order_items_data.dropna()
orders_data = orders_data.dropna()
sellers_data = sellers_data.dropna()

orders_data = orders_data[(orders_data['order_id'].isin(order_items_data['order_id']))].reset_index(drop= True)

# Taking English translation of product categories

products_data = products_data.merge(product_category_translation , on = 'product_category_name')
```

```python
## Converting from strings to Datetime objects

orders_data['order_purchase_timestamp'] = orders_data.apply(lambda row : datetime.strptime(row['order_purchase_timestamp'], '%Y-%m-%d %H:%M:%S'), axis = 1)
orders_data['order_approved_at'] = orders_data.apply(lambda row : datetime.strptime(row['order_approved_at'], '%Y-%m-%d %H:%M:%S'), axis = 1)
orders_data['order_delivered_carrier_date'] = orders_data.apply(lambda row : datetime.strptime(row['order_delivered_carrier_date'], '%Y-%m-%d %H:%M:%S'), axis = 1)
orders_data['order_delivered_customer_date'] = orders_data.apply(lambda row : datetime.strptime(row['order_delivered_customer_date'], '%Y-%m-%d %H:%M:%S'), axis = 1)
orders_data['order_estimated_delivery_date'] = orders_data.apply(lambda row : datetime.strptime(row['order_estimated_delivery_date'], '%Y-%m-%d %H:%M:%S'), axis = 1)


order_items_data['shipping_limit_date'] = order_items_data.apply(lambda row : datetime.strptime(row['shipping_limit_date'], '%Y-%m-%d %H:%M:%S'), axis = 1)
```

# Merging Datasets

```python
customer_loc = customers_data.merge(geolocation_data, left_on = 'customer_zip_code_prefix', right_on = 'geolocation_zip_code_prefix')

customer_loc = customer_loc.drop_duplicates(subset = 'customer_id', keep = False).reset_index(drop = True)

seller_location_data = sellers_data.merge(geolocation_data, left_on ="seller_zip_code_prefix", right_on = "geolocation_zip_code_prefix")#[['seller_id', 'geolocation_lat', 'geolocation_lng']]
Seller_loc = seller_location_data.drop_duplicates(subset = 'seller_id', keep = False).reset_index(drop = True)

```

```python
def hours_diff(row, col1, col2):
    hours = int(((row[col1] - row[col2]).total_seconds()/3600))
    return hours
```

```python
### Creating a few additional columns
orders_data['Approval_Time'] = orders_data.apply(lambda row : hours_diff(row, 'order_approved_at', 'order_purchase_timestamp') , axis = 1)
orders_data['Purchased_to_Shipped'] = orders_data.apply(lambda row: hours_diff(row, 'order_delivered_carrier_date', 'order_approved_at'), axis = 1)
orders_data['Shipped_to_Delivered'] = orders_data.apply(lambda row : hours_diff(row, 'order_estimated_delivery_date', 'order_delivered_carrier_date'), axis = 1)

orders_data['Hour'] = orders_data.apply(lambda row: row['order_purchase_timestamp'].hour, axis = 1)
orders_data['Weekday'] = orders_data.apply(lambda row: row['order_purchase_timestamp'].weekday(), axis = 1)
```

```python
order_items_time_data = orders_data.merge(order_items_data, on = 'order_id')

# Getting Seller Locations and delivery time in one place
order_items_time_seller_loc_data = order_items_time_data.merge(Seller_loc, on = 'seller_id')
```

# Patterns of orders by purchase in customers in terms of where are the orders coming from, general value of orders, what kind of service are they getting

```python
# Getting Customer Locations and delivery time in one place
order_items_time_customer_loc_data = order_items_time_data.merge(customer_loc, on = 'customer_id')
```

```python
# Saving the preprocessed variables (for multiple times closing and opening of the notebook and not having 
# to do the preprocessing steps again and again)

variables_dict = {
    'order_items_time_customer_loc_data' : order_items_time_customer_loc_data,
    'order_items_time_seller_loc_data' : order_items_time_seller_loc_data,
    'order_items_time_data' : order_items_time_data,
    'orders_data' : orders_data,
    'customers_data' : customers_data,
    'geolocation_data' : geolocation_data,
    'order_items_data' : order_items_data,
    'payments_data' : payments_data,
    'reviews_data' : reviews_data,
    'products_data' : products_data,
    'sellers_data' : sellers_data
}

pickle.dump(variables_dict, open('variables_dict.p', 'wb'))
```

## Location Clustering for customers

```python
customer_locations = order_items_time_customer_loc_data[['geolocation_lat', 'geolocation_lng']].values
```

```python
scanner = DBSCAN(eps=0.8, min_samples=2, metric='euclidean', algorithm='auto')
model = scanner.fit(customer_locations)
color_list = ['r', 'g', 'b', 'k', 'w']*1000
label_colors = [color_list[x] if x > 0 else 'y' for x in model.labels_]
plt.scatter(np.transpose(customer_locations)[0], np.transpose(customer_locations)[1], color = label_colors)
```

```python
Cluster_Series = pd.Series(data = model.labels_, name = 'Cluster_ID')
location_frame = order_items_time_customer_loc_data[['geolocation_lat', 'geolocation_lng']]
clustered_locations = pd.concat([Cluster_Series, location_frame], axis = 1)
```

```python
customer_group_locations = clustered_locations.groupby('Cluster_ID').mean().reset_index()
```

# Profit checking by frieght Value

## <u>Assumption : </u>
We are assuming a business model where price charged to the customer contains the freight_value

Earnings on a purchase are calculated by

Earnings = price - frieght_value

Sellers_earnings = 90% of Earnings

Websites_earnings = 10% of Earning which count as earnings for the website

```python
order_items_frieght = order_items_time_seller_loc_data[['product_id','seller_state','Shipped_to_Delivered', 'price', 'freight_value']]
```

```python
order_items_frieght['Website_Earnings'] = order_items_frieght.apply(lambda row: 0.1*(row['price'] - row['freight_value']), axis = 1)
order_items_frieght[['seller_state', 'Website_Earnings']].groupby('seller_state').sum().reset_index().plot.bar(x = 'seller_state', y = 'Website_Earnings')
```

```python
order_product_info = order_items_frieght.merge(products_data, on = 'product_id')
```

```python
order_product_info[['product_category_name_english', 'Website_Earnings']].groupby('product_category_name_english').sum().sort_values(by = 'Website_Earnings', ascending = False).reset_index().plot.bar(x = 'product_category_name_english', y = 'Website_Earnings')
```

```python
order_product_info[['product_category_name_english', 'Website_Earnings']].groupby('product_category_name_english').mean().sort_values(by = 'Website_Earnings', ascending = False).reset_index().plot.bar(x = 'product_category_name_english', y = 'Website_Earnings')
```
